# Push Changes and Create a GitHub Pull Request

This notebook demonstrates the process of preparing a local change, pushing a branch to GitHub, and opening a pull request.

## 1. Install and Configure Git

If Git is not installed, install it. Then configure your user name and email for commits.


````python
```bash
# Install Git if needed
sudo apt-get update && sudo apt-get install -y git

# Configure Git user metadata
git config --global user.name "Your Name"
git config --global user.email "you@example.com"
```
````

## 2. Clone Repository and Inspect Branch

Clone the target GitHub repository locally or navigate to the existing repo, then inspect the current branch and working tree.


````python
```bash
# Clone the repository if you do not already have it
# git clone https://github.com/Samuel1-ona/Hunty-contract.git
# cd Hunty-contract

# Inspect the current branch and status
git branch --show-current
git status --short --branch

# List files to confirm the working tree
ls -la
```
````

## 3. Create or Update Notebook File

Create a new notebook file or update an existing notebook in the repository to include the desired changes before committing.


````python
```bash
# Create or update the target notebook file inside the repository.
# For example, open the notebook in Jupyter or VS Code and make the desired changes.

# If this notebook is the change you want to commit, simply save it.
ls -la push_pr_demo.ipynb
```
````

## 4. Stage and Commit Changes

Stage your notebook changes and commit with a descriptive message.


````python
```bash
# Stage the notebook file
git add push_pr_demo.ipynb

# Commit with a descriptive message
git commit -m "Add notebook demonstrating how to push a branch and create a PR"
```
````

## 5. Push Branch to GitHub

Push your local branch to the GitHub remote and verify that the branch exists remotely.


````python
```bash
# Create a new branch or switch to the branch you want to push
git checkout -b fix/readme-github-issues-file

# Push the branch to the remote repository
git push --set-upstream origin fix/readme-github-issues-file

# Verify the remote branch exists
# git ls-remote --heads origin fix/readme-github-issues-file
```
````
